```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
movies = pd.read_csv('../../Data/movies_processed.csv', index_col=[0])
```

```python
movies.iloc[:,4:]
```

```python
movies['diversity_score'] = np.sum(movies.iloc[:,4:], axis=1)
```

```python
movies.diversity_score.mean()
```

```python
movies.diversity_score
```

```python
ratings = pd.read_csv('../../Data/ratings_processed.csv', index_col=[0])
```

```python
ratings.head()
```

```python
ratings = ratings.merge(movies[['movieId', 'diversity_score']], on='movieId')
```

```python
table = []
for usr in ratings.userId.unique():
    df = ratings[ratings.userId == usr]
    score = df.diversity_score.mean()
    table.append((usr, score))
```

```python
table = pd.DataFrame(table)
```

```python
table.columns = ['userId', 'averageDiversityofUser']
```

```python
table.describe()
```

```python
table.to_csv('../../Data/user_diversity.csv')
movies[['movieId', 'diversity_score']].to_csv('../../Data/movie_diversity.csv')
```
